### Univariate Modeling

In this notebook, we will try to find a univariate fit to predict our ETF. As we know, this ETF is highly volatile, so it is very unlikely that we find a good fit.

Before starting any actual coding, we are going to import all the required packages

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sktime.forecasting.all import temporal_train_test_split
from scipy import stats
from sklearn.metrics import mean_absolute_error as MAE
from sktime.performance_metrics.forecasting import mean_absolute_percentage_error
from sktime.performance_metrics.forecasting import MeanAbsolutePercentageError
from sklearn.metrics import mean_squared_error
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pmdarima as pm
from sktime.forecasting.arima import AutoARIMA
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import kpss, adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from prophet import Prophet
```

```python
# Pre defining recurrent functions
def is_stationary(series):
    display('Results of Dickey-Fuller Test:')
    adftest = adfuller(series, autolag='AIC')
    adf_output = pd.Series(adftest[0:4], index=['Test Statistic','p-value','Lags Used','Number of Observations Used'])
    for key,value in adftest[4].items():
        adf_output['Critical Value (%s)'%key] = value
    display(adf_output)
    
    display(f"Stationarity based of ADF Test: {adftest[1] <= 0.05}")
    
    display('Results of KPSS Test:')
    kpsstest = kpss(series, regression="ct") # I'm assuming stationarity around a trend. 
    kpss_output = pd.Series(kpsstest[0:3], index=['Test Statistic','p-value','Lags Used'])
    for key,value in kpsstest[3].items():
        kpss_output['Critical Value (%s)'%key] = value
    display(kpss_output)
    
    display(f"Stationarity based of KPSS Test: {kpsstest[1] >= 0.05}")
    
def smape_func(a, f):
    # source: https://www.statology.org/smape-python/
    return 1/len(a) * np.sum(2 * np.abs(f-a) / (np.abs(a) + np.abs(f))*100)

def mase_func(y_test, y_pred, y_train):
    # source: https://github.com/scikit-learn/scikit-learn/issues/18685
    e_t = y_test - y_pred
    scale = MAE(y_train[1:], y_train[:-1])
    return np.mean(np.abs(e_t / scale))

def evaluate_model(y_test, y_pred, y_train):
    # MAE 
    mae = MAE(y_test,y_pred)
    display(f"MAE: {mae}")
    
    # MSE
    mse = mean_squared_error(y_test, y_pred)
    display(f"MSE: {mse}")
    
    # MAPE
    mape = mean_absolute_percentage_error(y_test, y_pred)
    display(f"MAPE: {mape}")
    
    # SMAPE
    smape = smape_func(y_test, y_pred)
    display(f"sMAPE: {smape}")
    
    # MASE
    mase = mase_func(y_test, y_pred, y_train)
    display(f"MASE: {mase}")
```

### Loading and analyzing the series

```python
df = pd.read_csv("381170_index.csv")
df = df.rename(columns={'Date': 'ds'})
df = df.loc[:, ['ds', 'Price']]
df['Price'] = df['Price'].str.replace(',', '').astype(float)
df['ds'] = pd.to_datetime(df['ds'])
df['ds'] = df['ds'].dt.to_period('D')
df = df.set_index("ds")
df = df.sort_index(ascending=True)
df.head()
```

```python
# Summary
df.describe()
```

```python
# Single plot
plt.style.use("fivethirtyeight")
plt.figure(figsize=(8, 6))
plt.ylabel("Price")
plt.title("MiraeAsset TIGER US Tech Top10 INDXX")
df['Price'].plot(color='maroon')
plt.show()
```

```python
# Splitting data
y_train, y_test = temporal_train_test_split(df['Price'])
```

```python
# Checking for stationarity
is_stationary(y_train)
```

```python
# Checking for stationarity - Differenced data
is_stationary(y_train.diff()[1:])
```

```python
# Plotting differenced data
y_train.diff().plot(color='maroon')
```

```python
# Examine the autocorrelation plots - ACF
plot_acf(y_train)
plt.show()
```

```python
# Examine the autocorrelation plots - PACF
plot_pacf(y_train)
plt.show()
```

Given the decay in the ACF, and the droput in the PACF, we see a clear AR(1,1,0) order.

### ARIMA

```python
arima_model = sm.tsa.statespace.SARIMAX(endog=y_train,order=(1,1,0))
arima_fit = arima_model.fit()
display(arima_fit.summary())
```

```python
# Examining the residuals
arima_residuals = arima_fit.resid
# Plot
arima_residuals.plot()
```

```python
# Histogram
plt.hist(arima_residuals)
```

```python
# ACF
plot_acf(arima_residuals)
plt.show()
```

```python
# Ljung-Box
sm.stats.acorr_ljungbox(arima_residuals, return_df=True, boxpierce=True)
```

```python
# Predicting and plotting the predictions
y_pred = arima_fit.forecast(steps = len(y_test))
y_pred = pd.Series(y_pred)
y_pred.index = y_test.index
plt.figure(figsize=(12, 6))
y_train.plot(label='Train Data')
y_test.plot(label='Test Data')
y_pred.plot(label='Predicted Data')
plt.legend()
plt.title('ARIMA Model Prediction')
plt.show()
```

### AutoARIMA

```python
auto_arima = pm.auto_arima(y_train, D=1)
```

```python
display(auto_arima.arima_res_.summary())
```

```python
# Examining the residuals
auto_arima_residuals = auto_arima.arima_res_.resid
auto_arima_residuals.plot()
```

```python
# Histogram
plt.hist(auto_arima_residuals)
```

```python
# ACF
plot_acf(auto_arima_residuals)
plt.show()
```

```python
# Ljung-Box
sm.stats.acorr_ljungbox(auto_arima_residuals, return_df=True, boxpierce=True)
```

```python
# Predicting and plotting the predictions
y_pred = auto_arima.predict(n_periods=len(y_test))
y_pred = pd.Series(y_pred)
y_pred.index = y_test.index
plt.figure(figsize=(12, 6))
y_train.plot(label='Train Data')
y_test.plot(label='Test Data')
y_pred.plot(label='Predicted Data')
plt.legend()
plt.title('ARIMA Model Prediction')
plt.show()
```

### Prophet

```python
# Fitting vanilla Prophet
model = Prophet()
train_df = y_train.to_frame().reset_index()
train_df = train_df.rename(columns={'Price': 'y'})
train_df['ds'] = train_df['ds'].apply(lambda x: x.to_timestamp())
model.fit(train_df)
```

```python
# Predict
future_data = model.make_future_dataframe(periods=len(y_test), freq='d')
forecast_data = model.predict(future_data)
```
